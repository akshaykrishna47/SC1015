<b> not all values are non-null, need to fill in missing values and do data cleaning <b>

<b> DATA CLEANING <b>

```python
# check the number of null values 
df.isnull().sum()
```

```python
# for numerical values, we fill in the missing values with the mean 
df['IMDb-rating'] = df['IMDb-rating'].fillna(df['IMDb-rating'].mean())

# drop all other null values 
df = df.dropna()

df.info() 
df.head()
```
